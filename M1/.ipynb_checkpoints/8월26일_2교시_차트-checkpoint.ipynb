```python
plt.figure(figsize=(3,2))
X=np.linspace(-np.pi, np.pi, 256)
C=np.cos(X)

plt.title("x축과 y축의 tick label 설정")
plt.plot(X, C, label="테스트")
plt.xlabel("x축 이름")
plt.ylabel("y축 이름")

plt.xticks([-np.pi,  -np.pi/2,   0, np.pi/2,   np.pi])
plt.yticks([-1, 0, +1])

plt.xlim([-4.0, +4.0])
plt.ylim([-2.0, +2.0])

plt.legend(loc=0)
```

```python
X=np.linspace(-np.pi,np.pi,256)
C,S=np.cos(X),np.sin(X)
plt.title("legend를 표시한 플롯")
plt.plot(X,C,ls="--",label="cosine")
plt.plot(X,S,ls=":",label="sine")
plt.legend(loc=2)
plt.show()
```

```python
import numpy as np
import pandas as pd
from pandas import DataFrame
import matplotlib.pyplot as plt

# 2x2 서브플롯
fig = plt.figure()
ax3 = fig.add_subplot(2, 2, 3)
ax1 = fig.add_subplot(2, 2, 1) 
ax2 = fig.add_subplot(2, 2, 2)
ax4 = fig.add_subplot(2, 2, 4)

ax1.hist(np.random.randn(200), bins=20, color='k', rwidth=0.5)
ax2.scatter(np.arange(30), np.arange(30) + 3*np.random.randn(30))
ax3.plot(np.arange(10), np.random.randn(10))  
ax4.hist(np.random.randint(1, 7, 1000), bins=6, rwidth=1.0)  # width → rwidth

plt.show()

# 막대그래프용 데이터프레임
df = DataFrame(
    np.random.rand(6, 4),
    index=['one', 'two', 'three', 'four', 'five', 'six'],
    columns=pd.Index(['A', 'B', 'C', 'D'], name='Genus')
)

# 방법 A: 별도 Figure(기본 동작)
df.plot(kind='bar')
plt.show()

df.plot(kind='barh', stacked=True)
plt.show()

# --- 방법 B: 기존 fig의 특정 축에 그리고 싶다면 (예시) ---
# fig2, ax = plt.subplots()
# df.plot(kind='bar', ax=ax)
# plt.show()
```
