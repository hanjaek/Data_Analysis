# 탐색적 분석 (Exploratory Data Analysis, EDA)

- 데이터의 전반적인 특징을 파악하는 과정
- 분석에 앞서 수집한 데이터가 **적절한지 검증**
- 수집된 데이터의 전체적 특성을 확인하고, 기초 통계 및 시각화를 통해 인사이트 확보

---

## 주요 특징
- 기본적인 통계적 특성 파악  
  - 평균, 표준편차, 최소값, 최대값, 사분위수(25%, 50%, 75%) 등
- 데이터 시각화 활용  
  - 차트, 히스토그램, 박스 플롯(Box plot), 막대 그래프, 산점도(Scatter plot) 등

---

## 데이터 타입

### 1. 문자형 / 수치형 / 논리형 / 바이너리형
- **문자형**: 문자열 데이터 (예: 이름, 주소 등)  
- **수치형**: 수로 표현되는 데이터 (정수, 실수 등)  
- **논리형**: 참/거짓 (Boolean) 데이터  
- **바이너리형**: 비트 단위 데이터 (오디오, 비디오, 실행 파일 등)

### 2. 수치형 데이터의 세부 분류
- **범주형 (Categorical)**
  - 클래스 구분을 위한 값
  - 예: (월요일=1, 화요일=2 …), (남성=1, 여성=0)
  - 연산(+, -)에 의미 없음
- **순서형 (Ordinal)**
  - 순서만 의미가 있는 값
  - 예: 옷 사이즈(44, 55, 66…), 달력의 날짜(1일, 2일…)
  - 연산에 의미 없음
- **연속형 (Continuous)**
  - 수치 자체가 양적 의미를 가지는 값
  - 예: 무게, 길이, 온도, 압력, 속도, 화폐 단위
  - 덧셈/뺄셈 연산 결과가 의미 있음

---

## 데이터 형태

- **정형 데이터 (Structured Data)**
  - 포맷이 정해진 데이터
  - 예: Excel 표, CSV, XML, JSON 등

- **비정형 데이터 (Unstructured Data)**
  - 미리 정해진 포맷이 없는 데이터
  - 예: 블로그 글, 트위터 문장, 오디오/비디오

- **반정형 데이터 (Semi-Structured Data)**
  - 논리적 구조는 있으나, 완전한 정형 포맷은 아님
  - 예: 센서 데이터, 웹 사용 기록

---

## 데이터 수집 방법
1. **실제 측정** (센서, 관측 등)
2. **기존 데이터 활용** (기업 내부 데이터베이스 등)
3. **공공데이터** (개인정보 침해 범위 내에서 공개된 자료)
4. **파이썬에서 제공하는 샘플 데이터**
   - 진짜와 가짜 데이터가 혼합되어 있음
5. **캐글(Kaggle) 데이터셋**
   - 데이터 분석 대회용
   - 진짜와 가짜 데이터 혼재
